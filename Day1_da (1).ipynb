```python
#list count
#new list=[output for]
x=[78,32,37,64,14]
res=[i*i for i in x]
print(res)
```

```python
#list count
#newlist=[output for if]
x=[78,32,37,64,14]
res=[i*i for i in x if i>50]
print(res)
```

```python
#newlist=[(statement of if)if(condition)else(statement of else)for]
x=[78,32,37,64,14]
res=[i*i if i>50 else 'X' for i in x]
print(res)
```
