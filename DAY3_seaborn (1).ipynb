```python
import seaborn as sns

tit=sns.load_dataset('titanic')
print(tit)

sns.catplot(data=tit,y='class',x='age',kind='violin',hue='sibsp')

sns.catplot(data=tit,y='class',x='age',kind='box',hue='sibsp')

sns.catplot(data=tit,y='class',x='age',kind='swarm',hue='sibsp')

sns.catplot(data=tit,y='class',x='age',kind='bar',hue='sibsp')

iris=sns.load_dataset('iris')
print(iris)

sns.catplot(data=iris,y='petal_length',x='species',kind='violin',hue='sepal_width')

sns.catplot(data=iris,y='petal_length',x='species',kind='box',hue='sepal_width')

sns.catplot(data=iris,y='petal_length',x='species',kind='box',hue='sepal_width')

sns.catplot(data=iris,y='petal_length',x='species',kind='point',hue='sepal_width')

exe=sns.load_dataset('exercise')
print(exe)

sns.catplot(data=exe,y='id',x='time',kind='box',hue='kind')

sns.catplot(data=exe,y='id',x='time',kind='strip',hue='kind')

sns.catplot(data=exe,y='id',x='time',kind='point',hue='kind')
```
