```python
from bs4 import BeautifulSoup
```

```python
import requests as re
```

```python
x="https://www.geeksforgeeks.org/what-is-web-scraping-and-how-to-use-it/"
page=re.get(x)
```

```python
soup=BeautifulSoup(page.text,'html')
print(soup)
```

```python

#tittle
title=soup.find("title")
print(title)
```

```python
#title txt
print(title.text)
```

```python
y="https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1"
fobe=re.get(y)
```

```python
fobeindia=BeautifulSoup(fobe.text,"html")
print(fobeindia)
```

```python
#finding a table
table=fobeindia.find("table")
print(table)
```

```python
#finding a table head
tbhead=table.findAll("th")
print(tbhead)
```

```python
#it prints the main headings of the table
heading=[i.text for i in tbhead]
print(heading)
```

```python
import pandas as pd
#headinga are converted into the data frames
mydf=pd.DataFrame(columns=heading)
print(mydf)
```

```python
#print the rows
rows=table.findAll('tr')
print(rows)
```

```python
#print the table data
tabledata=table.findAll('td')
print(tabledata)
```

```python
data=[i.text for i in tabledata]
print(data)
```

```python
data[0:4]
data[4:8]
data[8:16]
```

```python
import pandas as pd
df=pd.DataFrame(columns=data)
print(df)
```

```python
ind=0
st=0
stop=4
for i in data:
  if stop <=40:
    mydf.loc[ind]=data[st:stop]
    st=st+4
    stop=stop+4
    ind=ind+1
print(mydf)
```
